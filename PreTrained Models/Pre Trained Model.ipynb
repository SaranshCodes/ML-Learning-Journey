```python
from keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
```

```python
# Load the VGG16 models and specify the input shape
model = VGG16(weights='imagenet' , include_top=True, input_shape=(224,224,3))
```

```python
import numpy as np
# Loading an image and preprocess it for the model
img_path=r"C:\Users\saran\Desktop\Shure_mikrofon_55S.jpg"
img= image.load_img(img_path, target_size = (224,224))
x=image.img_to_array(img)
x = np.expand_dims(x,axis=0)
x = preprocess_input(x)
```

```python
# USE THE MODEL TO PREDICT THE CLASS
pred= model.predict(x)
```

```python
pred
```

```python
# get the top5 predictions with class names
decode_preds = decode_predictions(pred,top=15)[0]
```

```python
for i in decode_preds:
    print(i[1],":",i[2])
```
