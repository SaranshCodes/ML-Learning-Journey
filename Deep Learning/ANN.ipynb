### ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initializing te ANN
classifier= Sequential()
# units: its an art, comes by experience. input+output/2
#initializer: how weights are updated
# relu activation function
# input_dim: input shape(shape of the data, how many features do we have in the data)
#adding the input layer and the first hidden layer
classifier.add(Dense(units = 6 , kernel_initializer='uniform', activation='relu',input_dim=10))
# Adding the second hidden layer
classifier.add (Dense(units=6, kernel_initializer='uniform', activation='relu'))

#  Adding the third hidden layer
classifier.add (Dense(units=6, kernel_initializer='uniform', activation='relu'))

#Adding the output
classifier.add (Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

# This entire code means firstly we have 10 neurons then there are 3 hidden layer which have 6 neurons each and then output layer. 
```

```python
len(x_train)
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
y_train[0]
# Fitting the ANN to the Training set
classifier.fit(x_train, y_train, batch_size = 10, epochs = 50)
```

```python
# making predictions and evaluating the model
y_pred_ann = classifier.predict(x_test)
y_pred_ann = (y_pred_ann >0.5)

cm3= confusion_matrix(y_test, y_pred_ann)
print(cm3)
```

```python
from sklearn.metrics import accuracy_score
score3 = accuracy_score(y_test, y_pred_ann)
print(score3*100)
```

```python
# improving the ANN classifier

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def build_classifier():
    model = Sequential()
    model.add(Dense(6, activation="relu", kernel_initializer="uniform", input_dim=10))
    model.add(Dense(6, activation="relu", kernel_initializer="uniform"))
    model.add(Dense(1, activation="sigmoid", kernel_initializer="uniform"))
    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model
classifier = KerasClassifier(
    model=build_classifier,
    verbose=0
)
pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler()),
    ("ann", classifier)
])
parameters = {
    "ann__batch_size": [10, 20],
    "ann__epochs": [5, 6]
}
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=parameters,
    scoring="accuracy",
    cv=5,
    n_jobs=-1
)

grid_search.fit(x_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(best_params)
print(best_score)
```
