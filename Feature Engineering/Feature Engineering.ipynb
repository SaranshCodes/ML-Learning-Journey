```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
df=pd.read_csv(r"C:\Users\saran\Desktop\Project\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.info()
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.dropna(how='any',inplace=True)
```

```python
x=df.drop(['customerID','Churn'],axis=1)
y=df['Churn']
```

```python
x
```

```python
y
```

```python
#convert categorical data into numericals ----> feature encoding
x=pd.get_dummies(x,columns=['gender', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],drop_first=True)
```

```python
#Split the data into train and test sets
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# Apply Recursive Feature Elimination (RFE) to select 15 best features

model=LogisticRegression()
rfe= RFE(model,n_features_to_select=15) # change here to get better accuracy
```

```python
# Fitting the model
rfe= rfe.fit(x_train,y_train)
```

```python
# Get the selected features

selected_features = x_train.columns[rfe.support_]
x_train_selected= x_train[selected_features]
x_test_selected= x_test[selected_features]
```

```python
# Build model
model_orig= LogisticRegression()
model_orig.fit(x_train,y_train)
y_pred_orig = model_orig.predict(x_test)

accuracy_orig = accuracy_score(y_test, y_pred_orig)
print("Accuracy of the base model is: ", round(accuracy_orig*100,2))
```

```python
# Build RFE model on 15 predictors
model_rfe= LogisticRegression()
model_rfe.fit(x_train_selected,y_train)
y_pred_rfe = model_rfe.predict(x_test_selected)

accuracy_rfe = accuracy_score(y_test, y_pred_rfe)
print("Accuracy of the RFE based model is: ", round(accuracy_rfe*100,2))
```

### SFS

```python
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
# Apply SFS to select 5 best features

model=LogisticRegression()
sfs= SequentialFeatureSelector(model,k_features=5, forward=True, floating= False, scoring= 'accuracy',cv=0) # change here to get better accuracy
```

```python
# Fitting the model
sfs = sfs.fit(x_train,y_train)
```

```python
# Get the selected features

selected_features = list(sfs.k_feature_names_)
x_train_selected= x_train[selected_features]
x_test_selected= x_test[selected_features]
```

```python
# Build RFE model on 5 predictors
model_sfs= LogisticRegression()
model_sfs.fit(x_train_selected,y_train)
y_pred_sfs = model_sfs.predict(x_test_selected)

accuracy_sfs = accuracy_score(y_test, y_pred_sfs)
print("Accuracy of the SFS based model is: ", round(accuracy_sfs*100,2))
```

### Chi Square

```python
from sklearn.feature_selection import chi2 ,SelectKBest
```

```python
# Apply chi2
chi2_selector= SelectKBest(chi2, k=20)
x_train_chi2 = chi2_selector.fit_transform(x_train,y_train)
x_test_chi2 = chi2_selector.transform ( x_test)
```

```python
# Build chi model on 5 predictors
model_chi2= LogisticRegression()
model_chi2.fit(x_train_chi2,y_train)
y_pred_chi2 = model_chi2.predict(x_test_chi2)

accuracy_chi2 = accuracy_score(y_test, y_pred_chi2)
print("Accuracy of the Chi2 based model is: ", round(accuracy_chi2*100,2))
```

### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=10)
x_train_pca= pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python
# Training the logistic regression model on training set

model_pca= LogisticRegression()
model_pca.fit(x_train_pca,y_train)
y_pred_pca = model_pca.predict(x_test_pca)

accuracy_pca = accuracy_score(y_test, y_pred_pca)
print("Accuracy of the PCA based model is: ", round(accuracy_pca*100,2))
```

### LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda= LinearDiscriminantAnalysis()
x_train_lda= lda.fit_transform(x_train,y_train)
x_test_lda=lda.transform(x_test)
```

```python
explained_variance = lda.explained_variance_ratio_
print(explained_variance)
```

```python
# Training the logistic regression model on training set

model_lda= LogisticRegression()
model_lda.fit(x_train_lda,y_train)
y_pred_lda = model_lda.predict(x_test_lda)

accuracy_lda = accuracy_score(y_test, y_pred_lda)
print("Accuracy of the LDA based model is: ", round(accuracy_lda*100,2))
```
